# Evaluation

## Objective

Evaluate how accurately ChatGPT detects errors in tabular datasets with different complexity levels.

## Input Files

* low_complex_test.csv → dataset containing injected errors
* low_complex_errors.csv → ground truth labels

## AI Evaluation Procedure

1. Upload the test dataset to ChatGPT.
2. Prompt ChatGPT to identify errors.
3. Record ChatGPT predictions.
4. Compare predictions with ground truth.
5. Measure detection performance.

## Prompt Format

You are given a tabular dataset containing possible data quality issues.

Task: Identify rows that contain errors and return:

row index
affected column
error type

Allowed error types:
Missing Value
Formula Error

Assumption:
Subtotal = Quantity × Price

Return results in table format.

Do not modify the dataset.
Only report detected errors.

## Evaluation Metrics

* Precision
* Recall
* F1 Score


# Experiment 1 = Low Complexity

## Model

ChatGPT

## Dataset

low_complex_test.csv

## Evaluation Criteria

A prediction is counted as correct only if all three conditions match the ground truth table:

* row number
* affected column
* error type

## Evaluation Metrics

The following metrics were used to evaluate model performance.

* True Positive (TP): Error correctly identified by the model.
* False Positive (FP): Error reported by the model that was not present in the ground truth.
* False Negative (FN): Existing error that was not detected by the model.

Metrics used:

* Precision = TP / (TP + FP)
  Measures how many predicted errors were correct.

* Recall = TP / (TP + FN)
  Measures how many actual errors were detected.

* F1 Score = harmonic mean of Precision and Recall
  Provides an overall measure of detection performance.
